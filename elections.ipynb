```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
elec = pd.read_csv('haryana2.csv')
elec.head()
```

```python
elec.isnull().sum()
```

```python
def value_cleaner(x):
    try:
        str_temp = (x.split('Rs')[1].split('\n')[0].strip())
        str_temp_2 = ''
        for i in str_temp.split(","):
            str_temp_2 = str_temp_2+i
        return str_temp_2
    except:
        x = 0
        return x
```

```python
elec['Total Assets'] = elec['Total Assets'].apply((value_cleaner))
elec['Liabilities'] = elec['Liabilities'].apply((value_cleaner))
elec.head()
```

```python
elec.dtypes
```

```python
elec['Total Assets']=pd.to_numeric(elec['Total Assets'])
elec['Liabilities']=pd.to_numeric(elec['Liabilities'])
elec['Criminal Case'].replace({np.NaN:0})
elec['Criminal Case'] = pd.to_numeric(elec['Criminal Case'], errors='coerce').fillna(0).astype(np.int64)
```

```python
elec_df=elec[elec['PARTY']!='NOTA']
elec_df['SEX'].replace({'MALE':1,'FEMALE':2},inplace=True)
elec_df['CATEGORY'].replace({'GENERAL':1,'SC':2},inplace=True)
i=1
parties_dict={}
for j in elec_df['PARTY']:
    if j in parties_dict:
        continue
    else:
        parties_dict[j]=i
        i+=1
elec_df['PARTY'].replace(parties_dict,inplace=True)
a=1
edu_dict={}
for b in elec_df['Education']:
    if b in edu_dict:
        continue
    else:
        edu_dict[b]=a
        a+=1
elec_df['Education'].replace(edu_dict,inplace=True)
df2 = elec_df[['AC NAME','Winner','PARTY','SYMBOL','SEX','Criminal Case','AGE','CATEGORY','Education','TOTAL','TOTAL ELECTORS','Total Assets','Liabilities']]
num_col = ['PARTY','Education','Criminal Case','AGE','TOTAL','TOTAL ELECTORS','Total Assets','CATEGORY','Liabilities','SEX']
dtst = pd.get_dummies(df2)
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
scaling_col = num_col
dtst[scaling_col] = standardScaler.fit_transform(dtst[scaling_col])
dtst.head()
```

```python
from sklearn.utils import resample
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
df_not_winner = dtst[dtst.Winner == 0]
df_winner = dtst[dtst.Winner == 1]
df_winner_upsampled = resample(df_winner, replace = True,n_samples = 1452, random_state = 0) 
df_total_upsampled = pd.concat([df_not_winner, df_winner_upsampled])
df_total_upsampled.Winner.value_counts()
y = df_total_upsampled['Winner']
X = df_total_upsampled.drop(['Winner'], axis = 1)

rf_scores = []
for k in range(1,60):
    randomforest_classifier= RandomForestClassifier(n_estimators=k,random_state=0)
    score=cross_val_score(randomforest_classifier,X,y,cv=10)
    rf_scores.append(score.mean())
```

```python
rf_scores
```

```python
randomforest_classifier= RandomForestClassifier(n_estimators=2,random_state=0)
score=cross_val_score(randomforest_classifier,X,y,cv=10)
print('% Accuracy :', round(score.mean()*100,4))
```
